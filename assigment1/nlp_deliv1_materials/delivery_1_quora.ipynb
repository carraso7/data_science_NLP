````python
def cast_list_as_strings(mylist):
    """
    return a list of strings
    """
    mylist_of_strings = []
    for x in mylist:
        mylist_of_strings.append(str(x))

    return mylist_of_strings
````

##  Extracting Features from Dataframe

Make a function `get_features_from_df` that given a dataframe containing the format of the training data
it returns a scipy sparse matrix with the features from question 1 and question 2 

````python
def get_features_from_df(df, count_vectorizer):
    """
    returns a sparse matrix containing the features built by the count vectorizer.
    Each row should contain features from question1 and question2.
    """
    q1_casted =  cast_list_as_strings(list(df["question1"]))
    q2_casted =  cast_list_as_strings(list(df["question2"]))
    
    ############### Begin exercise ###################
    # what is kaggle                  q1
    # What is the kaggle platform     q2
    X_q1 = count_vectorizer.transform(q1_casted)
    X_q2 = count_vectorizer.transform(q2_casted)    
    X_q1q2 = scipy.sparse.hstack((X_q1,X_q2))
    ############### End exercise ###################

    return X_q1q2
````

## Examining Mistakes

Make a function `get_mistakes` that given a model `clf` a dataframe `df`, the features `X_q1q2` and the target labels `y` it returns 


- incorrect_indices: coordinates where the model made a mistake
- predictions: predictions made by the model


````python
def get_mistakes(clf, X_q1q2, y):
    ############### Begin exercise ###################
    predictions = clf.predict(X_q1q2)
    incorrect_predictions = predictions != y 
    incorrect_indices,  = np.where(incorrect_predictions)
    ############### End exercise ###################
    
    if np.sum(incorrect_predictions)==0:
        print("no mistakes in this df")
    else:
        return incorrect_indices, predictions
````

## TFIDF VECTORIZER

````python
import scipy.sparse as sp
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. Initialize an improved Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    ngram_range=(1, 5),    # Use unigrams and bigrams
    min_df=3,              # Ignore terms that appear in less than 3 documents
    stop_words='english'   # Remove common stop words to reduce noise
)

# Fit on the previously created `all_questions` list to build the vocabulary
tfidf_vectorizer.fit(all_questions)

# 2. Build a new feature extraction function
def get_improved_features(df, vectorizer):
    """
    Extracts TF-IDF features and computes symmetric interaction matrices.
    """
    q1_casted = cast_list_as_strings(list(df["question1"]))
    q2_casted = cast_list_as_strings(list(df["question2"]))
    
    # Transform strings to TF-IDF sparse matrices
    X_q1 = vectorizer.transform(q1_casted)
    X_q2 = vectorizer.transform(q2_casted)
    
    # Compute symmetric interaction features
    # 1. Absolute difference highlights words unique to one question
    diff_features = abs(X_q1 - X_q2)
    
    # 2. Element-wise product highlights overlapping words
    prod_features = X_q1.multiply(X_q2)
    
    # Stack the interaction features horizontally
    X_combined = sp.hstack((diff_features, prod_features))
    
    return X_combined

# 3. Extract features for training and validation partitions
print("Extracting features...")
X_tr_improved = get_improved_features(train_df, tfidf_vectorizer)
X_val_improved = get_improved_features(val_df, tfidf_vectorizer)

# Extract targets
y_train = train_df["is_duplicate"].values
y_val = val_df["is_duplicate"].values

# 4. Train the Model
print("Training improved model...")
# Using a slightly higher max_iter to ensure convergence with new features
logistic_improved = LogisticRegression(solver="liblinear", random_state=123, max_iter=500)
logistic_improved.fit(X_tr_improved, y_train)

# 5. Evaluate the Model
print("Evaluating on Validation Data...")
y_pred_val = logistic_improved.predict(X_val_improved)

print(f"Validation Accuracy: {accuracy_score(y_val, y_pred_val):.4f}")
print("\nClassification Report:")
print(classification_report(y_val, y_pred_val))
````

````python
# Accuracy for improved TF-IDF model
X_te_improved = get_improved_features(test_df, tfidf_vectorizer)

train_acc_improved = accuracy_score(y_train, logistic_improved.predict(X_tr_improved))
val_acc_improved = accuracy_score(y_val, logistic_improved.predict(X_val_improved))
test_acc_improved = accuracy_score(y_test, logistic_improved.predict(X_te_improved))

print(f"Improved train accuracy: {train_acc_improved:.4f}")
print(f"Improved validation accuracy: {val_acc_improved:.4f}")
print(f"Improved test accuracy: {test_acc_improved:.4f}")
````

## length difference and word overlap ratio

````python
import numpy as np
import scipy.sparse as sp

def get_manual_features(df):
    """
    Calculates manual features: length difference and word overlap ratio.
    Returns a 2D numpy array of the features.
    """
    length_diffs = []
    overlap_ratios = []
    
    # Cast to strings using the function from your notebook
    q1_list = cast_list_as_strings(list(df["question1"]))
    q2_list = cast_list_as_strings(list(df["question2"]))
    
    for q1, q2 in zip(q1_list, q2_list):
        # 1. Length difference (character length difference)
        len_diff = abs(len(q1) - len(q2))
        length_diffs.append(len_diff)
        
        # 2. Word overlap ratio (using Jaccard Similarity: Intersection over Union)
        words_q1 = set(q1.lower().split())
        words_q2 = set(q2.lower().split())
        
        intersection = len(words_q1.intersection(words_q2))
        union = len(words_q1.union(words_q2))
        
        # Handle edge case where both strings are empty to avoid division by zero
        if union == 0:
            overlap_ratio = 0.0
        else:
            overlap_ratio = intersection / union
            
        overlap_ratios.append(overlap_ratio)
        
    # Stack the lists into a 2D array (shape: n_samples x 2)
    return np.column_stack((length_diffs, overlap_ratios))

def get_combined_features(df, vectorizer):
    """
    Combines the TF-IDF interaction features with the manual features.
    """
    # 1. Get the sparse TF-IDF interaction features (from the previous step)
    X_tfidf_symmetric = get_improved_features(df, vectorizer)
    
    # 2. Get the dense manual features
    X_manual = get_manual_features(df)
    
    # 3. Combine sparse and dense features using scipy.sparse.hstack
    # hstack automatically handles the conversion between dense and sparse formats
    X_combined = sp.hstack((X_tfidf_symmetric, X_manual))
    
    # Convert to Compressed Sparse Row (CSR) format for efficient modeling
    return X_combined.tocsr()

# --- Execution ---

print("Extracting combined features...")
X_tr_combined = get_combined_features(train_df, tfidf_vectorizer)
X_val_combined = get_combined_features(val_df, tfidf_vectorizer)

print("Training model with combined features...")
# Using a higher max_iter and scaling standardizes dense/sparse mix behavior
logistic_combined = LogisticRegression(solver="liblinear", random_state=123, max_iter=1000)
logistic_combined.fit(X_tr_combined, y_train)

print("Evaluating the new model...")
y_pred_val_combined = logistic_combined.predict(X_val_combined)

from sklearn.metrics import accuracy_score, classification_report
print(f"New Validation Accuracy: {accuracy_score(y_val, y_pred_val_combined):.4f}")
print("\nClassification Report:")
print(classification_report(y_val, y_pred_val_combined))
````

````python
# Accuracy for combined features model
X_te_combined = get_combined_features(test_df, tfidf_vectorizer)
train_acc_combined = accuracy_score(y_train, logistic_combined.predict(X_tr_combined))
val_acc_combined = accuracy_score(y_val, logistic_combined.predict(X_val_combined))
test_acc_combined = accuracy_score(y_test, logistic_combined.predict(X_te_combined))
print(f"Combined train accuracy: {train_acc_combined:.4f}")
print(f"Combined validation accuracy: {val_acc_combined:.4f}")
print(f"Combined test accuracy: {test_acc_combined:.4f}")

````
